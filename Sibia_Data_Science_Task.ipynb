```python
df1=df[['Customer Code','product_type_on_sales','Pack Classification','Pack Type','Pack Size','YTD Vol (MT)']]
```

```python
df1.head()
```

```python
x=df1['Pack Size']
x=x.to_list()
print(x)
```

```python
l=[]
for i in x:
    if i.endswith('KG'):
        y=i.replace('KG', '')
        y=int(y)
        y=y*1000
        l.append(y)
    elif i.endswith('LTR'):
        y=i.replace('LTR', '')
        y=int(y)
        y=y*1000
        l.append(y)
    elif i.endswith('GM'):
        y=i.replace('GM', '')
        y=int(y)
        l.append(y)
    elif i.endswith('ML'):
        y=i.replace('ML', '')
        y=int(y)
        l.append(y)
    else:
        y=int(i)
        l.append(i)
print(l) 
print(len(l))
```

```python
df1['pack_size(gm)']=l
```

```python
x = df1.drop(['YTD Vol (MT)'], axis = 1)
y = df1['YTD Vol (MT)']
```

```python
y
```

```python
x.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse=False,drop='first'), ['Pack Classification','Pack Type'])], remainder='passthrough')
x = ct.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler    
st= StandardScaler()    
x_train= st.fit_transform(x_train)    
x_test= st.transform(x_test)  
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)
```

```python
rf.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor 
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x_train, y_train)
```

```python
regressor.score(x_train, y_train)
```

```python
regressor.score(x_test, y_test)
```

```python
from sklearn import tree
tree.plot_tree(regressor,filled=True)
```
